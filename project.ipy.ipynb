```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel('/content/uncleaned_data.xlsx')
df=df1.copy()
```

```python
#1.	Identify and remove duplicate Item_ID rows
df.drop_duplicates(subset=['Item_ID'],keep='first',inplace=True)
df
```

```python
#2.	Identify duplicate production batch IDs
df['Batch_ID'].duplicated().value_counts()
```

```python
#3.	Fill missing Units_Produced with median per item
x=df['Units_Produced'].median()
df['Units_Produced']=df['Units_Produced'].fillna(x)
df['Units_Produced']
```

```python
#4.	Fill missing Defect_Count with 0 or median.
df['Defect_Count']=df['Defect_Count'].fillna(0)
df['Defect_Count']
```

```python
#5.	Fill missing Factory with "Unknown Factory
df['Factory']=df['Factory'].fillna('unknown factory')
df['Factory']
```

```python
#6.	Replace missing Supervisor with "Unassigned
df['Supervisor']=df['Supervisor'].fillna('unassigned')
df['Supervisor']
```

```python
#7.	Fill missing Production_Date with NaT or inferred date
df['Production_Date']=pd.to_datetime(df['Production_Date'],errors='coerce')
df['Production_Date']=df['Production_Date'].ffill()
df['Production_Date']
```

```python
#8.	Fill missing Machine_ID with "Unknown Machine".
df['Machine_ID']=df['Machine_ID'].fillna('unknown Machine')
df['Machine_ID']
```

```python
#9.	Replace missing Shift with "Unknown
df['Shift']=df['Shift'].fillna('unknown')
df['Shift']
```

```python
#10.	Fill missing Email with "noemail@factory.com
df['Email']=df['Email'].fillna('noemail@factory.com')
df['Email']
```

```python
#1.	Remove rows where Units_Produced <= 0
for i in df.index:
    if df.loc[i,'Units_Produced']<=0:
        df.drop(i,inplace=True)
df['Units_Produced']
```

```python
#2.	Remove rows with negative Defect_Count
for i in df.index:
    if df.loc[i,'Defect_Count']<=0:
        df.drop(i,inplace=True)
df['Defect_Count']
```

```python
#5.	Correct typos in Factory names.
df['Factory']=df['Factory'].replace({'Facotry C':'Factory C','factory a':'Factory A'})
df['Factory']
```

```python
#7.	Standardize Shift values (Morning/Afternoon/Night
df['Shift']=df['Shift'].replace({'Aft':'Afternoon','Morn':'Morning','night':'Night'})
df['Shift'].value_counts()
```

```python
#8.	Remove invalid Machine_ID formats
for i in df.index:
    if df.loc[i,'Machine_ID']=='MACH-XYZ':
        df.drop(i,inplace=True)
df['Machine_ID']
```

```python
#3.	Fetch runs in a specific factory
dept=int(input('choose  the factory:{Factory A:1, Factory B:2, Factory C :3}\n '))
if dept ==1:
  print(df[df['Factory']=='Factory A'])

elif dept ==2:
  print(df[df['Factory']=='Factory B'])

elif dept ==3:
  print(df[df['Factory']=='Factory C'])

elif dept ==4:
  print(df[df['Factory']=='unknown factory'])
else:
    print('invalid choice')
```
